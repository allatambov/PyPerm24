На первый взгляд, каждый элемент `votes_str` представляет собой полноценный словарь, просто заключённый в кавычки:

    '{ id:"160325", plus:"56", minus:"2", voted:"" }'

Однако это не так! Чтобы переделать строку в словарь, нужна строка, полностью соответствующая формату JSON, а здесь не хватает кавычек вокруг ключей. Как эту проблему решить? Найти по какому-то паттерну эти ключи и доклеить вокруг них кавычки с помощью функции `sub()` из модуля `re` для «умной» замены.

Функция `add_quotes()` для доклеивания кавычек уже написана, также ниже написан пример её использования вместе с `sub()`. Используя приведённую информацию, получите на основе списка `votes_str` список полноценных json-строк и сохраните его в переменную `votes_new`. Для самопроверки: вид строк в списке `votes_new` должен быть как в выдаче ниже рядом с `Стало:`.

```python
def add_quotes(x):
    """
    Parameters:
        x: match-object from re to be edited
    Returns:
        string with added quotes at the start and 
        at the end
    """
    # group() – из match извлекает саму подстроку (тип str)
    g = '"' + x.group() + '"'
    return g

# аргументы sub(): 
# pattern: шаблон, по которому ищем, что заменять
# repl: функция, которая выполняет замену
# string: строка, по который выполняем поиск

v = '{ id:"160325", plus:"56", minus:"2", voted:"" }'

print("Было:", v)
print("Стало:", re.sub(pattern = "[a-z]+", repl = add_quotes, string = v))
```
