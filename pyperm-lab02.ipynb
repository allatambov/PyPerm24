### Активировать тесты для задач с вводом данных

*Автор тестера – Щуров.И.В.*

```python
# Фабрика тестов для проверки программ, принимающих данные через input()

from collections import deque

class Tester(object):
    def __init__(self, inp):
        self.outputs = []
        self.inputs = deque(inp)
    def print(self, *args, sep = " ", end = "\n"):
        text = sep.join(map(str, args)) + end
        newlines = text.splitlines(keepends=True)
        if self.outputs and self.outputs[-1] and self.outputs[-1][-1] != "\n" and newlines:
            self.outputs[-1] += newlines[0]
            self.outputs.extend(newlines[1:])
        else:
            self.outputs.extend(newlines)
            
    def input(self, *args):
        assert self.inputs, "Вы пытаетесь считать больше элементов, чем предусмотрено условием"
        return self.inputs.popleft()
    def __enter__(self):
        global print
        global input
        print = self.print
        input = self.input
        return self.outputs
    def __exit__(self, *args):
        global print
        global input
        del print
        del input
```

### Вступление к задаче 6*

В **сетевом анализе** часто возникает необходимость фиксировать связи между людьми на основе их биографической информации. Так, иногда считается, что людей можно считать «связанными», то есть потенциально знакомыми, если они работали или учились вместе в одно и то же время. Данные о связях часто представляют в виде **матрицы смежности**. Это матрица размера $n \times n$, где $n$ – число людей, из 0 и 1, где 1 соответствует наличию связи между людьми, а 0 – её отсутствию. Пример матрицы смежности:

             Валя Вася Толя
    Валя       0   1   0
    Вася       1   0   0
    Толя       0   0   0
    
Это матрица соответствует случаю, если Вася и Валя знакомы друг с другом, при этом «знакомство» человека с самим собой не учитывается.

Для построения такой матрицы на основе реальных биографических данных в виде фрагментов резюме с местами работы/учёбы и соответствующими датами могут потребоваться следующие библиотеки, модули и функции.

1. Преобразовать строку с датой в определённом формате в специальный тип *дата-время* можно с помощью функций из [библиотеки](https://docs.python.org/3/library/datetime.html)`datetime`. Пример конвертации даты, содержащей только месяц и год:

        from datetime import datetime
        print(datetime.strptime('01.2023', '%m.%Y'))
    
        2023-01-01 00:00:00
        
2. Чтобы более эффективно и просто проверить пересечение временных интервалов, можно преобразовать результат в формате *дата-время* в строку типа `str`, а затем воспользоваться возможностями [библиотеки](https://datetimerange.readthedocs.io/en/latest/pages/examples/index.html) `datetimerange`. Для работы эту библиотеку необходимо установить, она не входит в базовый набор библиотек, устанавливаемых вместе с Anaconda. Пример проверки пересечения двух временных интервалов:

        from datetimerange import DateTimeRange
        range1 = DateTimeRange("2023-01-01 00:00:00", "2023-03-01 00:00:00")
        range2 = DateTimeRange("2023-02-25 00:00:00", "2023-04-01 00:00:00")
        range3 = DateTimeRange("2024-01-25 00:00:00", "2024-02-01 00:00:00")
        print(range1.is_intersection(range2))
        print(range1.is_intersection(range3))
        
        True
        False

3. Для нахождения пересечений двух списков с местами работы можно преобразовать списки в множества, это один из базовых типов данных в Python. Пример нахождения пересечения списков:

        A = [2, 7, 9, 0, 5]
        B = [2, 5, 3, 4]
        print(set(A).intersection(set(B)))

        {2, 5}
        
4. Для нахождения всех пар значений некоторого списка с повторениями можно воспользоваться функцией `product()` из модуля `itertools`, для нахождения всех уникальных пар значений можно воспользоваться функцией `combinations()` из этого же модуля. Примеры:

        from itertools import product, combinations
        print(list(product([1, 2, 3], repeat = 2)))
        print(list(combinations([1, 2, 3], r = 2)))
        
        [(1, 1), (1, 2), (1, 3), (2, 1), (2, 2), (2, 3), (3, 1), (3, 2), (3, 3)]
        [(1, 2), (1, 3), (2, 3)]
